#### RFE is a popular feature selection algorithm that predicts features that impact can target variables. RFE has wide variety of approaches that user can choose based on their needs. This model initially uses all the variables/features and appropriate methods are used to eliminate features that has less impact on the target variable. This method is recursively applied on the initial set of variables till the desired number of variables are reached. 

#### RFE is a part of sklearn library, and can be used for classificaiton and regression type of problems. 

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
crime_df=pd.read_csv('crime_data.csv')
crime_df
```

```python
X=crime_df.drop('X1',axis=1)
y=crime_df['X1']
```

```python
reg=LogisticRegression()
```

```python
selector = RFE(reg, n_features_to_select=4)
```

```python
rfe_sel=selector.fit(X,y)
```

```python
rfe_sel.support_
```

```python
rfe_sel.ranking_
```

### 2. Create a function called digital_root that takes in an integer. Digital root is the recursive sum of all the digits in a number.
### Given n, take the sum of the digits of n. If that value has more than one digit, continue reducing in this way until a single-digit number is produced. The input will be a non-negative integer.
Examples:
16 --> 1+6=7
942 --> 9+4+2=15 --> 1+5=6
132189 --> 1+3+2+1+8+9=24 --> 2+4=6 493193 --> 4+9+3+1+9+3=29 --> 2+9=11

```python
def digital_root (val):
    sum=0
    for i in (str(val)):
        sum=sum+int(i)
    if len(str(sum))==1:
        return sum
    else:
        sum2=0
        for i in (str(sum)):
            sum2=sum2+int(i)
        return sum2
```
