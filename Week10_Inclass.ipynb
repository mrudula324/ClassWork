### Dot Product of 2 vectors each can have n elements

```python
def dot_p(size):
    sum1=0
    print("Enter elements for Array1")
    m=[int(input()) for i in range(size)] 
    print("Array 1:",m) 
    print("Enter elements for Array2")
    n=[int(input()) for i in range(size)] 
    print("Array 2:",n)
    for i in range(size):
        m_d=(m[i]*n[i])
        sum1=sum1+m_d
    print ("Dot product of the Matrix:",sum1)
    return 
```

```python
dot_p(5)
```

### Cross Product for 3 elements given 2 vectors

```python
def cross_p(size): 
    s2=0
    print("Enter elements for Array1")
    m=[int(input()) for i in range(size)] 
    print("Array 1:",m) 
    print("Enter elements for Array1")
    n=[int(input()) for i in range(size)] 
    print("Array 2:",n)
    for i in range(size):
        s2=([m[1]*n[2] - m[2]*n[1],m[2]*n[0] - m[0]*n[2],m[0]*n[1] - m[1]*n[0]])
    print ("Cross product of the Matrix:",s2)
    return 
```

```python
cross_p(3)
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
path="belize_airbnb_listings.csv"
df=pd.read_csv(path)
df
```

```python
nei_gro=df['neighbourhood_group']
```

```python
uni_nei=df['neighbourhood_group'].unique()
```

```python
for nei,i in enumerate(uni_nei):
    print (nei,i)
```

```python
one_hot_en=[]
for x in nei_gro:
    if x =='Belize':
        temp=[1,0,0,0,0,0]
    elif x=='Cayo':
        temp=[0,0,0,1,0,0]
    elif x=='Stann Creek':
        temp=[0,1,0,0,0,0]
    elif x=='Toledo':
        temp=[0,0,1,0,0,0]
    elif x=='Corozal':
        temp=[0,0,0,0,1,0]
    else:
        temp=[0,0,0,0,0,1]
    one_hot_en.append(temp)
print(one_hot_en)
```
